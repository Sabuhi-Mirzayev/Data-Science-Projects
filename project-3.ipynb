```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv(r"C:\Users\User\Desktop\Datalar\musteriler.csv")
df=data.copy()
df
```

```python
df=df.dropna()
df=df[~df["InvoiceNo"].str.contains("C")]
df
```

```python
df["Country"].value_counts().sort_values(ascending=False)
```

```python
df=df[df["Country"]=="United Kingdom"]
```

```python
df.iloc[:,[0,2,3]]
```

```python
df=df.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
df
```

```python
def num(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
df=df.applymap(num)
```

```python
df1=apriori(df.astype("bool"),min_support=0.02,use_colnames=True)
df2=association_rules(df1,metric="confidence")
df2
```
